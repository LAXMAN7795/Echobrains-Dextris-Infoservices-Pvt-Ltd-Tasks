```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Display plots directly inside the notebook
%matplotlib inline
```

```python
# dataset from the sklearn
from sklearn.datasets import make_classification
```

```python
## create the dataset
X, y = make_classification(n_samples=1000, n_features=10,n_informative=3, n_classes=3, random_state=15)
# we get 3 classes 0,1,2
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
logistic=LogisticRegression(multi_class='ovr')# one versus rest(logistic regression type) for multi class classification
logistic.fit(X_train,y_train)
y_pred=logistic.predict(X_test)
```

```python
# metrics
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
score=accuracy_score(y_pred,y_test)
print(score)
print(classification_report(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
```
